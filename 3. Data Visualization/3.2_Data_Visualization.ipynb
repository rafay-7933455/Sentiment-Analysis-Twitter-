# Visualizing The Important Trends In The Cleaned Data

##### Loading The Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
```

##### Loading The Cleaned Data

```python
df = pd.read_csv('../data/cleaned.csv')
```

#### Pie Chart Representing Different Sentiments
###### Note: This Pie Chart Represents The Distribution Of Different Sentiments Within The Cleaned Data

```python
labels = df['Sentiment'].value_counts().index.to_list()
sizes = df['Sentiment'].value_counts().values
plt.pie(np.array(sizes), labels=labels, autopct='%1.1f%%')
plt.axis('equal')
plt.title('Overall Percentages of Different Sentiments')
plt.show()
```

#### Bar Plot Representing Overall Distribution
###### Note: This Simple Bar Plot represents the Overall Distribution of Cleaned Data in terms of Different Sentiments

```python
plt.bar(x=df['Sentiment'].value_counts().index, height=df['Sentiment'].value_counts().values.tolist(), color=['green','darkblue','red','yellowgreen'])
plt.title('Counts of Each Sentiment')
plt.show()
```

#### Bar Plot Representing Distribution Of Cleaned Data
###### Note: This BarPlot (i.e CountPlot) represents the Distribution of the Cleaned Data Among Different Tools in terms of Different Sentiments

```python
fig, axs = plt.subplots(1, 1, figsize=(10,6))
sb.countplot(data=df, x='Tool', hue='Sentiment', ax=axs)
axs.set_title('Multiple Barplot / CountPlot', fontsize=12)
axs.tick_params(axis='x', rotation=90)
axs.set_xlabel('Tool', fontsize=10)
axs.set_ylabel('No. Of Instances', fontsize=10)

fig.tight_layout()
plt.show()
```

##### Distribution in the Scatter Plot

```python
fig, axs = plt.subplots(1,1, figsize=(10,6))
sb.scatterplot(df, x='Tool', hue='Sentiment', y='Id', ax=axs)
axs.tick_params(axis='x', rotation=90)
axs.set_title('Id V/s Tool Distribtuion')
plt.legend()
plt.show()
```
