# Cleaning The Refined Data
__Steps Involved :__
1. Removing / Replacing _Null_ Values
2. Removing The _Duplicated_ Records
3. Saving The _Cleaned Data_

##### Loading Required Libraries

```python
import pandas as pd
```

##### Loading The Refined Data

```python
df = pd.read_csv('../data/refined.csv')
```

##### Removing Null Text Value Records

```python
print("Null Count:",df.isnull().sum(), sep='\n')
df.dropna(subset=['Text'], inplace=True)
print("Null Count:",df.isnull().sum(), sep='\n')
```

##### Removing Duplicated Records

```python
print("No. of Duplicated Records:", df.duplicated().sum())
dup = df.duplicated().sum()
df.drop_duplicates(inplace=True)
print(f"{dup} duplicates removed")
```

##### Saving the cleaned dataset

```python
df.to_csv('../data/cleaned.csv', index=False)
```
