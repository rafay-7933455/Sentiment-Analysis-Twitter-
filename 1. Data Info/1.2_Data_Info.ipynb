# Extracting Information From Raw Data

##### Loading The Required Libraries

```python
import pandas as pd
```

##### Loading The Raw File

```python
df = pd.read_csv('../data/twitter_training.csv')
```

##### Refining The Column Names
Note: The __Column Names__ were ambiguous and couldn't define the the data at all. Thus, the __Column Names__ need to be updated.

```python
df['Sentiment'] = df['Positive']
df['Id'] = df['2401']
df['Tool'] = df['Borderlands']
df['Text'] = df['im getting on borderlands and i will murder you all ,']
```

##### Removing Old Columns
Note: The **Data** of old columns is saved in the new columns defined above with better __Column Names__. Thus __Old Columns__ are being removing without any loss of __Data__

```python
df = df[['Sentiment','Id','Tool','Text']]
```

##### Saving the Refined data

```python
df.to_csv('../data/refined.csv', index=False)
```

##### No. of Records and Attributes

```python
print("No. of Records:",df.shape[0]) # Rows
print("No. of Attributes:",df.shape[1]) # Columns
```
